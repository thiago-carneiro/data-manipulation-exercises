### Tarefa 1. Importe o dataset e salve os dados em um dataframe

Os dados estão salvos no arquivo ***datasets/users_dataset.csv***.

Salve os dados em uma variável de nome *users*.

````python
import pandas as pd

users = pd.read_csv('https://raw.githubusercontent.com/thiago-carneiro/data-manipulation-exercises/master/datasets/users_dataset.csv',sep='|')
users.info()
````

### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
users.groupby('occupation').age.mean()
````

### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
pd.DataFrame(
                zip(
                    users.groupby('occupation').occupation.unique().apply(lambda x: x[0]),
                    users.groupby('occupation').age.min(),
                    users.groupby('occupation').age.max()
                    ),
             columns=['occupation','min_age','max_age'])

# ou tb:
# users.groupby('occupation').age.agg(['min','max'])
````

### Tarefa 4. Para cada combinação de ocupação e sexo, calcule a idade média.

Exemplo:

```
occupation     gender
administrator  F         40.638889
               M         37.162791
artist         F         30.307692
               M         32.333333
```

````python
users.groupby(['occupation','gender']).age.mean()
````

### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
count_users_by_occupation=users[['occupation','gender']].groupby('occupation').count()
count_users_by_occupation_gender=users[['occupation','gender','user_id']].groupby(['occupation','gender']).count()
count_users_by_occupation.rename({'gender':'contagem'},axis=1,inplace=True)
count_users_by_occupation_gender.rename({'user_id':'contagem'},axis=1,inplace=True)
count_users_by_occupation_gender/count_users_by_occupation
````
